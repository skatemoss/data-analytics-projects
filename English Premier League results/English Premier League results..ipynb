Футбол – один из самых популярных видов спорта на планете, за которым следят миллионы людей. В последние годы было собрано огромное количество данных о многих матчах в разных странах и лигах. Эти данные включают в себя информацию о каждом ударе или пасе, сделанном в матче.

Также, с недавнего времени, в футбольной индустрии большое распространение получила метрика xG (expected goals), которая оценивает вероятность забить гол в зависимости от местоположения игрока на поле и других факторов. Эта метрика позволяет более точно анализировать эффективность команд, а также оценивать игроков, их стоимость и потенциал импакта в команде.

4. **Команды с более высоким количеством желтых и красных карточек чаще теряют очки в матчах из-за снижения эффективности игры.**

5. **Команды с большим количеством атакующих действий вблизи ворот соперника (deep) чаще забивают голы и выигрывают матчи.**

6. **Команды, играющие на домашнем стадионе, побеждают чаще, чем в гостях.**

7. **Команды, которые наносят большее количество ударов в створ ворот чаще выигрывают.**

Возможно, существует влияние количества угловых ударов на результаты матчей для команд, играющих дома, и команд, играющих на выезде.

```python
# Создаем столбец 'winner'. 
def find_winner(row):
    if row['FTR'] == 'H':
        return row['homeTeam']
    elif row['FTR'] == 'A':
        return row['awayTeam']
    else:
        return 'Draw'

df['winner'] = df.apply(find_winner, axis=1)
```

```python
# Создаем два новых датафрейма, в которых содержатся только те матчи, в которых победила домашняя или гостевая команда соответственно.
home_games = df[df['homeTeam'] == df['winner']]
away_games = df[df['awayTeam'] == df['winner']]

# Вычисляем коэффициент корреляции между числом угловых ударов (HC) и количеством забитых голов для домашних и гостевых команд, которые выиграли матчи.
correlation_home = home_games['HC'].corr(home_games['homeGoals'])
correlation_away = away_games['AC'].corr(away_games['awayGoals'])

# Выводим на экран значения коэффициентов корреляции для домашних и гостевых команд соответственно.
print("Коэффициент корреляции для домашних матчей:", correlation_home)
print("Коэффициент корреляции для выездных матчей:", correlation_away)
```

**Выводы:**
* Коэффициент корреляции для домашних матчей составляет 0.0183, что указывает на очень слабую положительную связь между количеством угловых ударов и забитыми голами на домашнем стадионе. Это означает, что увеличение количества угловых ударов в домашних матчах слабо связано с увеличением количества забитых голов.


* Коэффициент корреляции для выездных матчей составляет -0.0047, что указывает на очень слабую отрицательную связь между количеством угловых ударов и забитыми голами на выезде. Это означает, что увеличение количества угловых ударов в выездных матчах практически не связано с изменением количества забитых голов.


* На основе этих результатов можно сделать вывод, что количество угловых ударов не имеет существенного влияния на результаты матчей, как для команд, играющих дома, так и для команд, играющих на выезде.

### Проверка гипотезы "команды, которые ведут в счете после первого тайма, чаще выигрывают матчи или набирают больше очков."

```python
# Создаем столбец для победителя первого тайма
df['ht_winner'] = df.apply(lambda row: row['homeTeam'] if row['HTHG'] > row['HTAG'] else (row['awayTeam'] if row['HTAG'] > row['HTHG'] else 'Draw'), axis=1)

# Создаем столбец для итогового победителя матча
df['final_winner'] = df.apply(lambda row: row['homeTeam'] if row['homeGoals'] > row['awayGoals'] else (row['awayTeam'] if row['awayGoals'] > row['homeGoals'] else 'Draw'), axis=1)

# Выбор матчей, где победитель первого тайма не "ничья"
non_draw_matches = df[df['ht_winner'] != 'Draw']

# Выбор матчей, где победитель первого тайма стал итоговым победителем
matches_won_by_ht_winner = non_draw_matches[non_draw_matches['ht_winner'] == non_draw_matches['final_winner']]

# Вычисление доли матчей, выигранных командой, ведущей после первого тайма
proportion_won_by_ht_winner = len(matches_won_by_ht_winner) / len(non_draw_matches)
print(f"Доля матчей, выигранных командой, ведущей после первого тайма: {proportion_won_by_ht_winner}")
```

```python
# Создаем датафрейм с необходимыми данными
data = {'Status': ['Won by HT winner', 'Not won by HT winner'],
        'Proportion': [proportion_won_by_ht_winner, 1 - proportion_won_by_ht_winner]}

pie_df = pd.DataFrame(data)

# Создаем круговую диаграмму
plt.figure(figsize=(8, 6))
colors = sns.color_palette("Set2")
plt.pie(pie_df['Proportion'], labels=pie_df['Status'], colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Proportion of matches won by the team leading after the first half')
plt.tight_layout()
plt.show()
```

**Вывод:**

* Доля матчей, выигранных командой, которая ведёт в счете после первого тайма составляет примерно 77.1%. **Исходя из этого, можно сделать вывод, что команды, которые ведут после первого тайма, действительно чаще выигрывают матчи.** 

### Проверка гипотезы "команды с более высоким количеством желтых и красных карточек чаще теряют очки в матчах из-за снижения эффективности игры"


```python
# Возьмем только домашние команды.
# Создаем новый столбец, который будет содержать 1, если домашняя команда выиграла, и 0 если команда проиграла.
df['home_win'] = (df['homeGoals'] > df['awayGoals']).astype(int)
```

```python
# Создаем новый столбец с общим количеством карточек, выданных в матче.
df['total_home_cards'] = df['HY'] + 2 * df['HR']
```

```python
# Разделим данные на обучающую и тестовую выборки.
X = df[['total_home_cards']]
y = df['home_win']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Обучим логистическую регрессию на обучающей выборке.
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# Оценим качество модели на тестовой выборке.
y_pred = logreg.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion matrix:", conf_matrix)
```

* Матрица ошибок (Confusion matrix) показывает следующие результаты:
 - 436 истинно-отрицательных предсказаний (True Negative, TN): модель правильно предсказала, что команда с большим количеством карточек проиграет.
 - 81 ложно-положительных предсказаний (False Positive, FP): модель ошибочно предсказала, что команда с большим количеством карточек выиграет.
 - 306 ложно-отрицательных предсказаний (False Negative, FN): модель ошибочно предсказала, что команда с большим количеством карточек проиграет.
 - 89 истинно-положительных предсказаний (True Positive, TP): модель правильно предсказала, что команда с большим количеством карточек выиграет.

```python
print("Logistic Regression coefficients:", logreg.coef_)
```

```python
# Создаем массив из возможного количества карточек
x_values = np.linspace(df['total_home_cards'].min(), df['total_home_cards'].max(), 100).reshape(-1, 1)

# Вычисляем предсказанные вероятности победы для каждого значения количества карточек
x_values_df = pd.DataFrame(x_values, columns=['total_home_cards'])
y_values = logreg.predict_proba(x_values_df)[:, 1]

# Строим график вероятности победы в зависимости от количества карточек
sns.lineplot(data=x_values_df, x='total_home_cards', y=y_values)
plt.title('Logistic Regression: Probability of home team winning vs. total cards')
plt.xlabel('Total cards')
plt.ylabel('Probability of home team winning')
plt.show()
```

**Выводы:**
* Точность модели составляет примерно 57.6%. Это означает, что модель правильно предсказывает исход матча (победа или поражение команды с большим количеством карточек) в 57.6% случаев. Учитывая, что точность бинарного классификатора составляет 50%, модель логистической регрессии незначительно превосходит бинарный классификатор.



* Коэффициенты логистической регрессии: -0.231. Отрицательный коэффициент указывает на то, что с увеличением количества карточек вероятность победы команды уменьшается. Однако коэффициент достаточно мал, что может говорить о слабом влиянии количества карточек на исход матча.


* **В целом, результаты анализа говорят о том, что влияние количества карточек на исход матча есть, но оно достаточно слабое.**
